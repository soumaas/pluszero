## 問題12

```python
img  =cv2.imread("imori.jpg")
W,H,C=img.shape

```

```python
ker_size=3
K       =np.diag([1]*ker_size).astype("float")
K       =K/ker_size
```

```python
#ゼロパディング
pad         = ker_size // 2
out         = np.zeros((H + pad * 2, W + pad * 2, C), dtype=np.float)
out[:H,:W]  = img.copy().astype(np.float)
tmp         = out.copy()
```

```python
for y in range(H):
  for x in range(W):
    for c in range(C):
      out[y,x,c]=np.sum(K*tmp[y:y+ker_size,x:x+ker_size,c])
```

```python
out = out[: H,: W].astype(np.uint8)
```

```python
plt.imshow(out)
```

## 問題13

```python
img  =cv2.imread("imori.jpg")
W,H,C=img.shape
ker_size=3
out=img.copy()
```

```python
for h in range(H):
  for w in range(W):
    for c in range(C):
      out[h,w,c]=np.max(img[h:h+ker_size,w:w+ker_size,c])-np.min(img[h:h+ker_size,w:w+ker_size,c])
```

```python
out = out[: H,:W].astype(np.uint8)
plt.imshow(out)
```

```python

def Histogram_normalization(X_in,a,b,c,d):
  list=[]
  for x_in in X_in:
    if x_in < c:
      x_out = a
      list.append(x_out)

    elif x_in >= c and x_in < d:
      x_out = (b - a)*(x_in - c)/(d - c)+ a
      list.append(x_out)

    else:
      x_out = b
      list.append(x_out)

  return list
```

```python
#画像読込
img = cv2.imread("imori.jpg")

# 画像サイズ
height = img.shape[0]  # 高さ
width  = img.shape[1]  # 幅

#画像の回転行列
rot_matrix = cv2.getRotationMatrix2D((width/2,height/2),  # 回転の中心座標
                                      180,                # 回転する角度
                                      1,                  # 画像のスケール
                                    )

# アファイン変換適用
afin_img = cv2.warpAffine(img,               # 入力画像
                          rot_matrix,        # 行列
                          (width,height)     # 解像度
                         )
```
